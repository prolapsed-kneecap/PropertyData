```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/train.csv')
data
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(data.full_sq, data.price_doc, alpha=0.2)
plt.xlabel("Стоимость")
plt.ylabel("Общая площадь")
```

```python
model = LinearRegression()
```

```python
m1 = data['full_sq'].mean()
m2 = data['price_doc'].mean()
data['full_sq'] = data['full_sq'].fillna(m1)
data['price_doc'] = data['price_doc'].fillna(m2)
x = pd.DataFrame(data.full_sq)
y = pd.DataFrame(data.price_doc)
```

```python
model.fit(x, y)
```

```python
model.coef_
```

```python
model.intercept_
```
